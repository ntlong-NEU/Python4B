```python
text_input = '''ChatGPT is    an AI assistant.\n\n It understands natural language.\n\n It helps answer    questions fast.\n\n It can assist with writing \n\ntasks.\n\n ChatGPT learns    from vast data.\n\nIt adapts to    user needs well.\n\n This AI     is by OpenAI's team.\n\n It's built to aid     and inform.\n\n Ask ChatGPT anything today!'''
text_input_nor = text_input.replace('  ',' ')
text_input_nor = text_input_nor.replace('\n',' ')
text_input_nor = text_input_nor.replace('  ',' ')

print(text_input_nor)
```

```python
max_size = 63
end_points = []
for i in range(len(text_input_nor)):
    if text_input_nor[i] in {'.','!','?',';'}:
        end_points.append(i)
print(end_points)
```

```python
start = 0
end = 0
cuts = []
for i in range(len(end_points)):
    if end_points[i]> end +max_size:
        end = end_points[i-1]
        cuts.append((start,end))
        start = end 

cuts.append((start,len(text_input_nor)))
cuts
```

```python
pars  = [text_input_nor[start: end] for start, end in cuts]
print(pars)
```
