### 🟢 Mức độ Dễ (4 Bài)

**Bài 1: Tính doanh thu cơ bản**
* **Yêu cầu:** Viết hàm `calculate_revenue(price, quantity)` nhận vào giá bán của một sản phẩm và số lượng bán ra. Hàm trả về tổng doanh thu.

**Bài 3: Kiểm tra chỉ tiêu đạt được**

**Bài 4: Trả về nhiều giá trị (Tuple)**
* **Yêu cầu:** Viết hàm `convert_currency(usd_amount)` nhận vào số tiền USD. Hàm trả về đồng thời 2 giá trị: số tiền quy đổi ra VND (tỷ giá 25,000) và số tiền quy đổi ra EUR (tỷ giá 0.92).

```python
def convert_currency(usd_amount):
    vnd = usd_amount * 25000
    eur = usd_amount * 0.92
    return vnd, eur

# Kiểm tra
vnd_amount, eur_amount = convert_currency(100)
print(f"100 USD = {vnd_amount} VND, {eur_amount} EUR")
```

### 🟡 Mức độ Trung bình (4 Bài)

**Bài 5: Tính tổng chi phí linh hoạt với `*args`**
* **Yêu cầu:** Viết hàm `calculate_total_cost(*args)` sử dụng tham số độ dài biến thiên để nhận vào một số lượng không xác định các khoản chi phí phát sinh trong tháng. Hàm trả về tổng của tất cả các khoản chi phí đó.

```python
def calculate_total_cost(*args):
    total = 0
    for cost in args:
        total += cost
    return total # Hoặc đơn giản là: return sum(args)

# Kiểm tra
print("Tổng chi phí:", calculate_total_cost(150, 200, 50, 400))
```

**Bài 6: Lọc danh sách khách hàng tiềm năng**
* **Yêu cầu:** Viết hàm `filter_customers(spending_list, min_spending)` nhận vào một list (danh sách) các giá trị chi tiêu của khách hàng và một mức tối thiểu. Hàm sử dụng vòng lặp để tạo và trả về một list mới chỉ chứa các khoản chi tiêu lớn hơn hoặc bằng mức tối thiểu.

```python
def filter_customers(spending_list, min_spending):
    potential_customers = []
    for spending in spending_list:
        if spending >= min_spending:
            potential_customers.append(spending)
    return potential_customers

# Kiểm tra
spending_data = [50, 120, 300, 45, 80, 250]
print("KH chi tiêu cao:", filter_customers(spending_data, 100))
```

**Bài 7: Thống kê mô tả cơ bản từ một dãy số**
* **Yêu cầu:** Viết hàm `get_sales_statistics(sales_list)` nhận vào một list các số thực. Hàm tính toán và trả về một kiểu dữ liệu `Dictionary` chứa 3 khóa: `"Tổng"`, `"Cao nhất"`, `"Trung bình"`.

```python
def get_sales_statistics(sales_list):
    if not sales_list: # Xử lý trường hợp list rỗng
        return None
    total = sum(sales_list)
    max_sales = max(sales_list)
    average = total / len(sales_list)
    return {
        "Tổng": total,
        "Cao nhất": max_sales,
        "Trung bình": round(average, 2)
    }

# Kiểm tra
sales_data = [12.5, 15.0, 10.2, 22.4, 18.1]
print("Thống kê:", get_sales_statistics(sales_data))
```

**Bài 8: Hàm ẩn danh (Lambda) và hàm Map**
* **Yêu cầu:** Cho một danh sách giá gốc của các sản phẩm. Hãy dùng hàm `map()` kết hợp với một hàm `lambda` để tạo ra một danh sách giá mới sau khi đã cộng thêm 10% thuế VAT.

```python
base_prices = [1000, 2500, 500, 3000]

# Sử dụng lambda và map
prices_after_tax = list(map(lambda x: x * 1.10, base_prices))

# Kiểm tra
print("Giá sau thuế:", prices_after_tax)
```

### 🔴 Mức độ Khó (2 Bài)

**Bài 9: Viết hàm đệ quy (Recursion) tính lãi kép**
* **Yêu cầu:** Viết hàm đệ quy `calculate_compound_interest(principal, rate, years)` để tính tổng số tiền nhận được (cả gốc lẫn lãi) sau `years` gửi ngân hàng. Công thức kỳ tiếp theo bằng tiền kỳ trước cộng thêm `tiền kỳ trước * rate`. *Lưu ý: Phải dùng đệ quy, không dùng vòng lặp hay công thức lũy thừa đóng.*

```python
def calculate_compound_interest(principal, rate, years):
    # Điều kiện dừng: Khi số năm = 0, tiền giữ nguyên
    if years == 0:
        return principal
    
    # Tính tiền gốc cho năm tiếp theo (bao gồm cả gốc và lãi của năm hiện tại)
    new_principal = principal + (principal * rate)
    
    # Gọi đệ quy giảm dần số năm
    return calculate_compound_interest(new_principal, rate, years - 1)

# Kiểm tra (Gửi 100 triệu, lãi 6%/năm, trong 3 năm)
print("Tổng tiền sau 3 năm:", round(calculate_compound_interest(100, 0.06, 3), 2))
```

**Bài 10: Xây dựng Decorator kiểm tra tính hợp lệ của dữ liệu đầu vào**
* **Yêu cầu:** Viết một decorator có tên `@validate_number_input`. Decorator này sẽ bọc các hàm tính toán, với nhiệm vụ kiểm tra xem tất cả các tham số `*args` truyền vào có phải là kiểu số (`int` hoặc `float`) hay không. Nếu có bất kỳ đối số nào là chuỗi (string) hoặc kiểu khác, in ra "Lỗi: Dữ liệu đầu vào phải là số!" và không thực thi hàm. Nếu hợp lệ, trả về kết quả bình thường.

```python
def validate_number_input(func):
    def wrapper(*args):
        # Kiểm tra từng đối số xem có phải là int hoặc float không
        for arg in args:
            if not isinstance(arg, (int, float)):
                print(f"Lỗi: '{arg}' không phải là số hợp lệ!")
                return None
        # Nếu tất cả đều là số, tiến hành gọi hàm gốc
        return func(*args)
    return wrapper

# Áp dụng decorator vào hàm tính tổng lợi nhuận
@validate_number_input
def calculate_total_profit(*args):
    return sum(args)

# Kiểm tra
print("Trường hợp đúng:", calculate_total_profit(100, 250, 300))
print("Trường hợp sai:")
calculate_total_profit(100, "Hai trăm", 300)
```
