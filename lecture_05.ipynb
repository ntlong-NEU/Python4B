```python
def text_normalize(text:str)->str:
    """normalize text"""
    text_norm = text.lower()
    special_chars = set('.,-*!?;')
    for char in special_chars:
        text_norm = text_norm.replace(char, ' ')
    return text_norm
```

```python
def count_fre(text:str, normalize = True)->dict:
    if normalize:
        text_norm  = text_normalize(text)
    else:
        text_norm = text 
    fre_dict = {}
    for w in text_norm.split():
        if w in fre_dict:
            fre_dict[w] +=1
        else:
            fre_dict[w] = 1
    return fre_dict

    
```

```python
count_fre(zen_python)
```

```python
count_fre(zen_python, normalize=False)
```
