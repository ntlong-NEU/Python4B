1. **Chuẩn bị Dữ Liệu**:
   - Tải dữ liệu từ GitHub bao gồm các tệp: `state-population.csv`, `state-areas.csv`, và `state-abbrevs.csv`.
   - Đọc dữ liệu vào các DataFrame bằng `pd.read_csv`.

2. **Hiển thị Dữ Liệu**:
   - Hiển thị dữ liệu đầu của các tệp để kiểm tra cấu trúc và nội dung.

4. **Xử lý Dữ Liệu Thiếu**:

```python
df1 = pd.DataFrame({'key': list("ABCABC"),
                    "data1":range(6),
                    'data2':range(6,12)},
                    columns=['key','data1','data2'])
df1 
```

```python
df1.groupby('key').sum()
```

```python
df1.groupby('key').aggregate(['sum','min'])
```

```python
def myfunc(x):
    return x.mean()//2
```

```python
df1.groupby('key').aggregate(['sum','min',myfunc])
```

```python
df1.groupby('key').agg({'data1':['sum'], "data2":myfunc})
```
