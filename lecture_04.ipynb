## Lecture 4: Conditional Statements and While loops

```python
n  = 1
while n<5:
    print(n)
    n += 1
```

```python
for i in range(5):
    for j in range(5):
        print(i,' ',j, 'i+j=', i+j)
        if i+j == 2:
            break
```
